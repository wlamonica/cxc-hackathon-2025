```python
df = df.melt(id_vars = ['Country Name', 'Country Code', 'Indicator Name', 'Topic'], value_vars=years, var_name='Year', value_name='val')
```

```python
df = df.pivot_table(columns='Indicator Name', index =['Country Name', 'Year'], values='val').reset_index()
```

```python
nac_table = country_table.isna()
nac_table.head()
```

```python
keep_columns = []

for i in nac_table.columns:
    if nac_table[i].sum() < 35:
        keep_columns.append(i)
    print(i + ": " + str(nac_table[i].sum()))
```
